# Objects in Programming
1.1 General Definition
    
Objects in computer science can be variables, functions, data structures or anything else which *is a location in memory and has a value assigned to it, which can be called upon with a command*.
    
1.2 Objects in Python
    
Python is an [object oriented programming language](https://en.wikipedia.org/wiki/Object-oriented_programming). Object oriented programming is a design praxis to refer to reusable modules and systems. Object oriented programming languages try to integrate *code* and *data* into a so-called **object**. An object is an abstract concept with a **behaviour** (code) and a **state** (data). A practical example of an object would be the tab "shoes" on Zalando. In addition to objects, Python assigns classes and subclasses to objects. Classes and its subclasses assign hierarchical order to behaviours and states. Some plausible subclasses for the "shoes" object on Zalando might be "sports shoes", "sneakers" or "hiking boots". It is possible to define your own classes and subclasses, however, for entry level Python this will likely not be needed. If you want to find out how to play around with classes, click [here](https://en.wikibooks.org/wiki/A_Beginner%27s_Python_Tutorial/Classes), if you are more interested in the theoretical background of object oriented Python, click [here](https://docs.python.org/2/tutorial/classes.html) for the official Python.org documentation.


# Lists

In Python, a list is a very flexible data structure which has a wide range of built-in functions to your use. A list's two most important properties are that their elements are both *changeable* and in *ordered sequences*. You can spot lists by their **[square brackets]**. Lists are a great tool for when you work with many values, want to track a certain value's development or if you want to perform the same operation on a large number of different values. Defining an empty list in Python is really simple.
    
<code>sample_list = []</code>
    
Different elements in lists are separated by commas:
    
<code>sample_list = [1, 2, 3]</code>
    
Some commands to work with lists are listed below:
    
<code>sample_list.append(x)</code>  Adds an item to the end of the list.
     
<code>sample_list.insert(i, x)</code>  Insert an item at a given position. The first argument is the index of the element *before which to insert*, the second argument is the element. 
     
<code>sample_list.count(x)</code>  Counts the number of times that *x* appears in the list.
     
Note that the syntax always follows the same principles:
     
<code>listname.command(argument(s))</code>
     
A more detailed list of commands can be found [here](https://docs.python.org/3/tutorial/datastructures.html). For a more intuitive guide to Python's list function, consult this [blogpost](https://www.digitalocean.com/community/tutorials/understanding-lists-in-python-3).
     
## Sets

A concept closely related to lists are sets. Sets have two striking differences to lists in Python: First, they consist of **unique elements** and are an **unordered** sequence. As it is unordered, sets do not support indexing, slicing or other ordered-sequence operation. Its uses are found in the removal of duplicates or similar. For more on sets, refer to the [Python Library](https://docs.python.org/2/library/sets.html).
    

# Formatting the <code>print</code> outputs
    
Next, we will cover how to change the way the <code>print</code> command displays elements on your output. Namely, we are interested in font colour, as well as how to do bold statements. Interestingly, *italic* output is not supported by standard Python. Arnau used the following line in his [Three Queens Code](https://fronter.com/unisg/links/link.phtml?idesc=1&iid=1078955) (version 4):
    
<code>print('\033[1;31mYour final balance is {:.3f} CHF\033[1;31m'.format(money))</code>

To call the graphics function, you start by typing this argument <code>\033[</code>. The next value you enter specifies the *text attributes*. After separating with a semicolon <code>;</code>, the next value specifies the *colour*. After the colour code, you end the argument with a lowercase <code>m</code>. Note that this function remains active until the next occurence of the sequence, hence Arnau putting it at the end of the <code>print</code> statement again. Let's start with a list of some useful text attributes:
    
<code>0 = All attributes off
1 = Bold on
4 = Underscore on (not always supported)</code>
    
Our sample argument if we want to print in bold now looks like this:
    
<code>\033[1;</code> 
    
Now, let's list some colours, given by double-digit arguments:
    
<code>30 = Black
31 = Red
32 = Green
33 = Yellow
34 = Blue
35 = Magenta
36 = Cyan
37 = White</code>
    
A sample statement could be:
    
<code>print('\033[1;36mThis string is bold and Cyan\033[1;36m')</code>
        
Arnau also uses another handy feature in his <code>print</code> command:
    
<code>print('\033[1;31mYour final balance is {:.3f} CHF\033[1;31m'.format(money))</code>
    
Notice the <code>{:.3f}</code>. Since our Three Queens code will possibly spit out very small floats with a lot of decimals, Arnau implements this instead of the usual <code>{}</code> to display exactly three decimals. Intuitively, the integer in the expression specifies the number of decimals. A sample code that makes use of this could be:

<code>1: n = 12.124436457
2: print(f'look at this nice rounding {n:.4f}')
</code>

# The <code>random</code> function
    
Lastly, we spoke about the <code>random</code> function. As <code>random</code> is not a built-in function of Python, the first thing you need to do in order to make use of it, is to import <code>numpy</code>. <code>random</code> is used, as indicated by the name, to generate pseudo-random numbers, on demand also according to different distributions such as the standard normal or others. There exists the possibility of seeding the RNG (*Random Numbers Generator*) to synchronise the outcome of the <code>random</code> function (see box).

<div class="alert alert-block alert-info"> 
On the importance and meaning of seeding:
<br><br>
The random function in python does not generate actual random numbers - it generates *pseudo* - random numbers. What this means is that the computer performs operations on a given value to generate a number, the next number using the previously generated number. However, the first time you use the generator, there is no initial value. (Side remark: Since Python generates values at runtime, when you re-run the code, it will again lack an initial value!) Python works around this by using the current time on your computer as first value if you do not specify a seed. 
<br><br>
*Seeding* a pseudo - random numbers generator gives it its first "previous" value. Each seed value will generate the same sequence of pseudo - random numbers on each machine. That is, if Arnau any you type in the same seed, you will get the same sequence of outcomes! More on this will be discussed in next week's lecture. 
<br><br>
You can provide a seed through the command <code>random.seed()</code>
</div>
    
Now, to go back to the <code>random</code> function, the feature that we can most handily use when wanting to generate integers for our Three Queens Game is the function <code>random.randint()</code>. This can either be used by giving one parameter:

<code>random.randint(b)</code> which will generate numbers from 0 to b-1.

Alternatively, use two parameters to indicate the range: <code>random.randint(a, b)</code>. **Caution**: Unlike other instances where you indicate a range, here the upperbound <code>b</code> is included! This function generates numbers from <code>a</code> to <code>b</code>.

For a more extensive and detailed description of <code>random</code>, consult the [Python Library](https://docs.python.org/3/library/random.html)
